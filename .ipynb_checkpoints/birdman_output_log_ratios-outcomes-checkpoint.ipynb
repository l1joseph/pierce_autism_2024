```python
import pandas as pd
import numpy as np
```

```python
birdman_tables = {
    'dementia': pd.read_csv('/home/lakhatib/Wisconsin_MARS/metagenomics/birdman/diagnosis/normal_dementia/species/inferences-results/normal_dementia_birdman_l7_ft.beta_var.tsv', sep="\t"),
    'amyloid': pd.read_csv('/home/lakhatib/Wisconsin_MARS/metagenomics/birdman/amyloid/species/inferences-results/amyloid_birdman_l7_ft.beta_var.tsv', sep="\t"),
    'apoe4': pd.read_csv('/home/lakhatib/Wisconsin_MARS/metagenomics/birdman/apoe4/species/inferences-results/apoe4_birdman_l7_ft.beta_var.tsv', sep="\t"),
}

bottom_features_list = {}
top_features_list = {}
# Extract lower and upper bounds of the confidence interval
for key in birdman_tables.keys():
    if 'dementia' in key:
        test = 'diagnosis[T.Normal]'
    elif 'amyloid' in key:
        test = 'amyloid[T.Positive]'
    else:
        test = 'apoe4[T.Positive]'
   
    birdman_tables[key][['lower_bound', 'upper_bound']] = birdman_tables[key][f'{test}_hdi'].str.strip('()').str.split(',', expand=True).astype(float)
    # Check if the interval does not contain 0
    birdman_tables[key]['significant_interval'] = (birdman_tables[key]['lower_bound'] > 0) | (birdman_tables[key]['upper_bound'] < 0)

    # Keep only the rows with significant intervals
    birdman_tables[key] = birdman_tables[key][birdman_tables[key]['significant_interval']]

    # Extract species name from Feature column
    birdman_tables[key]['Feature'] = [col.split(';')[-1][3:] for col in birdman_tables[key]['Feature']]
    
    if 'dementia' in key:
        # Reverse values
        birdman_tables[key] = birdman_tables[key].copy()
        birdman_tables[key][f'{test}_mean'] = -birdman_tables[key][f'{test}_mean']
        birdman_tables[key]['lower_bound'] = -birdman_tables[key]['lower_bound']
        birdman_tables[key]['upper_bound'] = -birdman_tables[key]['upper_bound']
        
    # Extract features with positive mean
    bottom_features_list[key] = birdman_tables[key][birdman_tables[key][f'{test}_mean'] < 0]['Feature'].tolist()

    # Extract features with negative mean
    top_features_list[key] = birdman_tables[key][birdman_tables[key][f'{test}_mean'] > 0]['Feature'].tolist()


```

```python
#import feature table and metadata
import qiime2 as q2
ft = q2.Artifact.load('/home/l1joseph/pierce_autism_2024/data/wolr2/zebra30/195330_zebra.qza').view(pd.DataFrame)
md = pd.read_csv('/home/l1joseph/pierce_autism_2024/data/metadata/Metadata_Final.tsv', sep='\t', index_col=0)
```

```python
#extract species name
ft.columns = [col.split(';')[-1][3:] for col in ft.columns]
```

```python
#sum top and bottom feature abundance
for key in bottom_features_list.keys():
    ft[f'bottom_{key}'] = ft[bottom_features_list[key]].sum(axis=1)
for key in top_features_list.keys():
    ft[f'top_{key}'] = ft[top_features_list[key]].sum(axis=1)
```

```python
#calculate log ratio
for key in birdman_tables.keys():
    ft[f'log_ratios_{key}'] = np.log(ft[f'top_{key}'] / ft[f'bottom_{key}'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import kruskal

# Set up the figure layout
fig, axes = plt.subplots(3, 3, figsize=(15, 13))  # Adjust the figsize as needed

# Define common settings
jitter_value = 0.1  # Jitter for the stripplot
dot_size = 5  # Dot size for the stripplot

# Plot each dataset
datasets = ['log_ratios_dementia', 'log_ratios_amyloid', 'log_ratios_apoe4']
titles = ['Diagnosis', 'Amyloid', 'APOE4']
groups = ['diagnosis', 'amyloid', 'apoe4']
palette = [['lightblue', 'darkblue'], ['lightcoral', 'darkred'], ['lightgreen', 'darkgreen']]

for i, dataset in enumerate(datasets):
    for j, group in enumerate(groups):
        ax = axes[i][j]
        test_col = dataset
        group_col = group
        ft_test = ft.dropna(subset=group)
        group1_data = ft_test[ft_test[group_col] == ft_test[group_col].unique()[0]][test_col]
        group2_data = ft_test[ft_test[group_col] == ft_test[group_col].unique()[1]][test_col]

        # Perform the Kruskal-Wallis test
        statistic, pvalue = kruskal(group1_data, group2_data)

        # Boxplot
        sns.boxplot(x=group_col, y=test_col, data=ft_test, ax=ax, palette=palette[j], width=0.5)
        # Stripplot
        sns.stripplot(x=group_col, y=test_col, data=ft_test, ax=ax, color='k', size=dot_size, jitter=jitter_value, alpha=0.5, edgecolor='gray', linewidth=0.5)

        # Setting titles and texts
        if group_col == 'apoe4':
            group_title = group_col.upper()
        else:
            group_title = group_col.capitalize()
        ax.set_title(f'{titles[i]} Features by {group_title}', fontsize=14)
        p_value_text = f"< .001" if pvalue < 0.001 else f"{pvalue:.2f}"
        ax.text(0.5, .06, f'Kruskal-Wallis: {statistic:.2f}\n p-value: {p_value_text}', ha='center', va='center', fontsize=10, transform=ax.transAxes)
        ax.set_xlabel(group_title, fontsize=12)
        ax.set_ylabel(f"log(Top {titles[i]}/Bottom {titles[i]})", fontsize=12)
        
            # Setting the linewidth and color of the axes spines
        for spine in ax.spines.values():
            spine.set_linewidth(1)  # Adjust this value to change the thickness of the border
            spine.set_edgecolor('black') 
            
        subplot_labels = ['A)', 'B)', 'C)', 'D)', 'E)', 'F)', 'G)', 'H)', 'I)']
        for ax, label in zip(axes.flat, subplot_labels):
            # Position the text in the top-left corner of each subplot
            ax.text(-0.08, 1.07, label, transform=ax.transAxes, fontsize=16, va='top', ha='right')
# Tight layout to ensure no overlap
plt.tight_layout()
plt.savefig('/home/lakhatib/Wisconsin_MARS/metagenomics/birdman/png_results/log_ratios_birdman.png', dpi=600)  # Adjust the path and DPI as necessary
plt.show()
```

```python
# Define a function to handle column summation
def sum_columns_ignore_missing(df, column_list):
    valid_columns = [col for col in column_list if col in df.columns]
    if not valid_columns:
        print(f"No valid columns found in {column_list}")
        return None
    return df[valid_columns].sum(axis=1)

#Perform summation for each category
#sum top and bottom feature abundance
for key in bottom_features_list.keys():
    adrc_ft[f'bottom_{key}'] = sum_columns_ignore_missing(adrc_ft, bottom_features_list[key])
    
for key in top_features_list.keys():
    adrc_ft[f'top_{key}'] = sum_columns_ignore_missing(adrc_ft, top_features_list[key])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import kruskal

# Set up the figure layout
fig, axes = plt.subplots(1, 2, figsize=(10, 5))  # Adjust the figsize as needed

# Define common settings
jitter_value = 0.1  # Jitter for the stripplot
dot_size = 5  # Dot size for the stripplot

# Plot each dataset
tables = {'Present Cohort': ft, 'Validation Cohort': adrc_ft}

for j, key in enumerate(tables.keys()):
    ax = axes[j]
    test_col = 'log_ratios_dementia'
    group_col = 'diagnosis'
    ft_test = tables[key].dropna(subset=[test_col, group_col])
    group1_data = ft_test[ft_test[group_col] == ft_test[group_col].unique()[0]][test_col]
    group2_data = ft_test[ft_test[group_col] == ft_test[group_col].unique()[1]][test_col]

    # Perform the Kruskal-Wallis test
    statistic, pvalue = kruskal(group1_data, group2_data)

    # Boxplot
    sns.boxplot(x=group_col, y=test_col, data=ft_test, ax=ax, width=0.5, palette = ['lightblue', 'darkblue'])
    # Stripplot
    sns.stripplot(x=group_col, y=test_col, data=ft_test, ax=ax, color='k', size=dot_size, jitter=jitter_value, alpha=0.5, edgecolor='gray', linewidth=0.5)

    ax.set_title(f'{key}', fontsize=14)
    p_value_text = f"< .001" if pvalue < 0.001 else f"{pvalue:.2f}"
    ax.text(0.5, .06, f'Kruskal-Wallis: {statistic:.2f}\n p-value: {p_value_text}', ha='center', va='center', fontsize=10, transform=ax.transAxes)
    ax.set_xlabel("Diagnosis", fontsize=12)
    ax.set_ylabel(f"log(Top Dementia/Bottom Dementia)", fontsize=12)

        # Setting the linewidth and color of the axes spines
    for spine in ax.spines.values():
        spine.set_linewidth(1)  # Adjust this value to change the thickness of the border
        spine.set_edgecolor('black') 

    subplot_labels = ['A)', 'B)']
    for ax, label in zip(axes.flat, subplot_labels):
        # Position the text in the top-left corner of each subplot
        ax.text(-0.08, 1.07, label, transform=ax.transAxes, fontsize=16, va='top', ha='right')
# Tight layout to ensure no overlap
plt.tight_layout()
plt.savefig('/home/lakhatib/Wisconsin_MARS/metagenomics/birdman/png_results/adrc_validation_log_ratios_birdman.png', dpi=600)  # Adjust the path and DPI as necessary
plt.show()
```
